# Singapore HDB Resale Analytics: Exploratory Data Analysis

This notebook explores the processed HDB resale flat data, visualizes trends, and provides geospatial insights.

```python
# Import required libraries
import pandas as pd
import plotly.express as px
import geopandas as gpd
import os
```

## 1. Load Processed Data
Load the cleaned HDB resale data from the processed parquet file.

```python
# Load processed data
proc_path = '../data/processed/hdb_resale_processed.parquet'
df = pd.read_parquet(proc_path)
df.head()
```

## 2. Basic Data Overview
Show info, describe, and check for missing values.

```python
# Data info and summary
df.info()
```

```python
# Describe statistics
df.describe(include='all')
```

```python
# Check missing values
df.isnull().sum()
```

## 3. Price Trends Over Time
Visualize median resale price trends over time.

```python
# Median resale price by month
trend = df.groupby('month')['resale_price'].median().reset_index()
px.line(trend, x='month', y='resale_price', title='Median HDB Resale Price Over Time')
```

## 4. Price by Town and Flat Type
Visualize median price by town and by flat type.

```python
# Median price by town
by_town = df.groupby('town')['resale_price'].median().sort_values().reset_index()
px.bar(by_town, x='resale_price', y='town', orientation='h', title='Median Resale Price by Town')
```

```python
# Median price by flat type
by_flat = df.groupby('flat_type')['resale_price'].median().sort_values().reset_index()
px.bar(by_flat, x='flat_type', y='resale_price', title='Median Resale Price by Flat Type')
```

## 5. Geospatial Visualization
If geo data is available, plot a choropleth map of median price by town.

```python
# Plot choropleth if geojson exists
geojson_path = '../data/geo/singapore_planning_areas.geojson'
if os.path.exists(geojson_path):
    gdf = gpd.read_file(geojson_path)
    town_price = df.groupby('town')['resale_price'].median().reset_index()
    fig = px.choropleth_mapbox(
        town_price,
        geojson=gdf,
        locations='town',
        featureidkey='properties.PLN_AREA_N',
        color='resale_price',
        center={"lat": 1.3521, "lon": 103.8198},
        mapbox_style='open-street-map',
        zoom=10,
        title='Median Resale Price by Town (Choropleth)'
    )
    fig.show()
else:
    print('GeoJSON not found, skipping map.')
```
